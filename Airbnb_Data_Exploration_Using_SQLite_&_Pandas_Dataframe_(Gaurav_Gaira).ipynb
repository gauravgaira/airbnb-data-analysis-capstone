```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
df = pd.read_csv('/content/Airbnb NYC 2019 (2).csv')
```

```python
df.head()
```

```python
#connect to a sqllite database
conn = sqlite3.connect('airbnb_database.db')
#('airbnb_database.db') database name
```

```python
#now lets make an table and import all data in sqlite3
# if exist mean if before table hai to replace and also index false mean index ko load mat kro wrna ek extra column add
df.to_sql('airbnb_table', conn, if_exists='replace', index=False)
```

```python
# if we want to write sql query in python so
query = 'SELECT * FROM airbnb_table'
#but to read we have to write pd.read_sql_query(whatever query, connection name )
pd.read_sql_query(query, conn)
```

```python
query = 'SELECT * FROM airbnb_table'
#but to read we have to write pd.read_sql_query(whatever query, connection name )
pd.read_sql(query, conn)
```

# where room type = Private room

```python
#now lets write a query where room type = Private room
query = 'SELECT * FROM airbnb_table where room_type = "Private room" '
pd.read_sql_query(query, conn)
```

# query to find neighbourhood = Harlem

```python
#write a sql query to find neighbourhood -=Harlem
query = 'SELECT * FROM airbnb_table where neighbourhood = "Harlem" '
pd.read_sql_query(query, conn)
```

# all details with availability_365 exist

```python
#find all details with availability_365 exist
query = 'SELECT count(*) FROM airbnb_table where availability_365 = 365 '
pd.read_sql(query, conn)
```

##**CREATING TABLE IN SQLITE BY HELP OF PYTHON**

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('employee1.db')
```

```python
# create table
conn.execute('''create table employee_data
(emp_id int primary key,
 emp_name varchar(50) not null,
 dept varchar(15),
company varchar(20) default 'Almabetter',
salary decimal not null ,
exp decimal check (exp > 2))''')
```

```python
# insert records
conn.execute("""insert into employee_data
	values(101,'Jeevitha','IT','Almabetter',90000,3)""")
```

```python
conn.execute('''insert into employee_data
values(102,'Ankita','HR','Almabetter',100000,4),
(103,'Shivam','Sales','Almabetter',120000,6),
(104,'Vikas','Sales','Almabetter',120000,6)''')
```

```python
var=conn.execute("select * from employee_data")
```

```python
pd.read_sql("select * from employee_data",conn)
```
